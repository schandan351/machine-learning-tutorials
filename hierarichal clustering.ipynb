```python
# importing packgaes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('machine learning/K_Means/K_Means/Mall_Customers.csv')
dataset.head(10)
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
# lets plot our dendogram to find the no of clusters
# discuss about dendogram
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("dendrogram")
plt.xlabel("no of clusters")
plt.ylabel("euclidian distance")
plt.show()
```

```python
# so from dendrogram we choose no of clusters by looking into largest vertical distance 
# we choose such that without crossing any horizontal line
```

```python
# lets import hierarichal clustering
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
# visualizing our datasets
plt.scatter(X[y_hc ==0,0],X[y_hc==0,1],s=100,c='red',label='cluster1')
plt.scatter(X[y_hc ==1,0],X[y_hc==1,1],s=100,c='blue',label='cluster1')
plt.scatter(X[y_hc ==2,0],X[y_hc==2,1],s=100,c='cyan',label='cluster1')
plt.scatter(X[y_hc ==3,0],X[y_hc==3,1],s=100,c='magenta',label='cluster1')
plt.scatter(X[y_hc ==4,0],X[y_hc==4,1],s=100,c='green',label='cluster1')
plt.title("clustering the datasets")
plt.xlabel("income")
plt.ylabel("customers point")
plt.legend(loc='upper right')
plt.show()
```
